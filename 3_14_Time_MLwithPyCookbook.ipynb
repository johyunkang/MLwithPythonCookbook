## 3.14 시간에 따라 행을 그룹핑하기

```python
import pandas as pd
import numpy as np

# 날짜 범위를 만든다
time_index = pd.date_range('2023-01-02', periods=100_000, freq='30s', tz='Asia/Seoul')
print(time_index[:10])
print('\n\ntime_index 길이:', len(time_index))

# df 만들기
df = pd.DataFrame(index=time_index)

# 난숫값으로 열을 만듭니다.
df['sale_amount'] = np.random.randint(1, 10, 100_000)

display(df.head())


print('\n\n1주 단위로 행을 그룹핑한 다음 합을 계산')
display(df.resample('w').sum())


print('\n\n2주 단위로 행을 그룹핑한 다음 평균을 계산')
display(df.resample('2w').mean())

print('\n\한 달 단위로 행을 그룹핑한 다음 카운트를 계산')
display(df.resample('m').count())

```

위 결과를 보면 각 각 주와 월 단위 그룹핑했지만 `datetime` 인덱스는 날짜임.
기본적으로 `resample` 메서드는 시간 그룹핑의 오른쪽(right) 에지(edge)의 레이블(마지막 레이블)을 반환함. 하지만 그 기준을 변경 가능

```python
print('그룹핑의 첫 날짜로 날짜 인덱스 만들기')
display(df.resample('m', label='left').count())

print('\n\n월의 시작날짜로 만들기')
display(df.resample('MS').count()) # MS: Month Start, M : Month End

# 참조 URL : http://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
```
