## 3.18 데이터프레임 연결하기

```python
import pandas as pd

data_a = {'id':['1', '2', '3'],
          'first':['alex', 'amy', 'allen'],
          'last':['anderson', 'ackerman', 'ali']}
data_b = {'id':['4', '5', '6'],
          'first':['길동', '심청', '이유'],
          'last':['홍', '정', '아']}

df_columns=['id', 'first', 'last']
df_a = pd.DataFrame(data_a, columns=df_columns)
df_b = pd.DataFrame(data_b, columns=df_columns)

display(df_a)
display(df_b)
```

```python
# 행 방향(아래로) DF를 연결 (UNION ALL과 동일)
print('행방향 (아래로) DF를 연결 (UNION ALL과 동일)')
pd.concat([df_a, df_b], axis=0)
```

```python
print('열방향 (옆으로) DF를 연결 (JOIN 과 동일)')
pd.concat([df_a, df_b], axis=1)
```

## 3.19 데이터프레임 병합하기 (merge)

```python
import pandas as pd

emp_data = {'emp_id':[1, 2, 3, 4],
            'name':['아이유', '안유진', '전소연', '김사나']}
sales_data = {'emp_id':[3, 4, 5, 6],
              'total_sales':[1000, 2000, 3000, 4000]}
df_emp = pd.DataFrame(emp_data)
df_sales = pd.DataFrame(sales_data)

display(df_emp)
display(df_sales)
```

```python
print('데이터프레임을 병합(join) 합니다.')
df_merge = pd.merge(df_emp, df_sales, on='emp_id')
display(df_merge)

print('\n\n')
print('Outer Join (how=outer)')
display(pd.merge(df_emp, df_sales, on='emp_id', how='outer'))

print('\n\n')
print('Left Outer Join (how=left)')
display(pd.merge(df_emp, df_sales, on='emp_id', how='left'))


print('\n\n')
print('조인을 위한 컬럼명이 다른 경우 left_on, right_on')
display(pd.merge(df_emp, df_sales, left_on='emp_id', right_on='emp_id'))
```
